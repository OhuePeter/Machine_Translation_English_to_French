```python
import gradio as gr    # UI library
from transformers import pipeline   # Transformer pipeline
```

### Load Pipeline

```python
translation_pipeline = pipeline('translation_en_to_fr')
```

```python
translation_pipeline("It is raining cats and dogs")
```

```python
# To get the translated text:
text_t = translation_pipeline("It is raining cats and dogs")
```

```python
text_t
```

```python
text_t[0]['translation_text']
```

### Create Gradio Function and Interface

```python
# Create a new function called translate:
def translate_transformers(from_text):
  text_t = translation_pipeline(from_text)
  return text_t[0]['translation_text']
```

```python
translate_transformers("Today is Sunday")
```

```python
# Wrap this inside our gradio app:
interface = gr.Interface(fn=translate_transformers,
                        inputs=gr.inputs.Textbox(lines=2, placeholder='Text to translate'),
                         outputs='text')
```

```python
interface.launch()
```
